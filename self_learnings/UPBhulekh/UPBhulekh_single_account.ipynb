```python
import pytesseract
import time
import cv2
import easyocr
from PIL import Image
import re


# def read_captcha(image_path):
#     captcha_image = Image.open(image_path)
#     captcha_text = pytesseract.image_to_string(captcha_image)
#     return captcha_text.strip()
def read_captcha(captcha_image_path):


    image = cv2.imread(captcha_image_path)
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    reader = easyocr.Reader(['en'])
    result = reader.readtext(gray)
    # Extract the recognized text
    captcha_text = ""
    for res in result:
        captcha_text += res[1] + " "

    captcha_text =  captcha_text.strip()
    captcha_text = captcha_text[0::2].strip()
    print("captcha_text-----",captcha_text)

    return captcha_text


def fill_captcha_and_continue(driver, captcha_text):
#     captcha_input  =result_page.find_element(By.CSS_SELECTOR,"#imgCaptcha").screenshot('captcha.png')
#     captcha_input.clear()
    input_textArea = driver.find_element(By.XPATH,'//*[@id="cap"]').send_keys(captcha_text)

    try:
        button_continue = driver.find_element(By.XPATH,'//*[@id="capdiv"]/button/b')
        button_continue.click()
    except:
        pass
    
try:    

    def main():

        flag = True


        while flag:
    #         driver.get("https://upbhulekh.gov.in/public/public_ror/action/capImage.jsp")

            # Wait for the CAPTCHA image to load
            captcha_image  =WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"#imgCaptcha")))
            captcha_image.screenshot('captcha.png')

            captcha_text = read_captcha('captcha.png')
            print("CAPTCHA Text:", captcha_text)

            pattern = r'[!@#$%^&*()_+{}\[\]:;"\'<>,.?\\|]'

#             if re.search(pattern,captcha_text) !=True: 
#                 print("special characters are not present") 
            
            if (len(captcha_text) == 6 )and (not re.search(pattern,captcha_text)):
                fill_captcha_and_continue(driver, captcha_text)
                try:
                    success_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="mdn-container"]/form/table/tbody[1]/tr[1]/th/center/div/label[1]')))
                    failure_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/center/main/div[3]/div[1]/div/div[1]/ul/li/span' )))
                    if "खाता विवरण (अप्रमाणित प्रति)" in success_element.text:
                        print("Success! CAPTCHA is correct.")
                        flag = False
                        break
                    else:
                        print("captcha not matched")
    #                     if "Captcha Not Matched" in failure_element:
    #                         print("CAPTCHA is incorrect. Trying again...")                    

                        refresh_btn = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="ref-captcha"]/div/i'))).click()
        #                 captcha_image  =WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"#imgCaptcha")))
        #                 captcha_image.screenshot('captcha.png')
        #                 captcha_text = read_captcha('captcha.png')

                        time.sleep(2)
            
            # Adding a delay before retrying
                except TimeoutException:
                    pass
                    
                
            else:
                print("CAPTCHA length is not 6 or special characters present. Trying again...")
                refresh_btn = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="ref-captcha"]/div/i'))).click()

                time.sleep(2)  # Adding a delay before retrying
#             else: 
#                 refresh_btn = driver.find_element(By.XPATH,'//*[@id="ref-captcha"]/div/i').click()
#                 print("clicking refresh button")
#                 time.sleep(2)


        driver.quit()
        
except TimeoutException :
    pass
    
    
    
    
    

if __name__ == "__main__":
    try:
        main()
    except:
        pass
```
